```python
import pandas as pd
import plotly.express as px
```

### Convex Hull Benchmarks
1) `n = 20`, random points

2) `10 < n < 1000`, random points

```python
df = pd.read_csv("convexhull_01.csv")
df = df.iloc[:100]  # Plot first 100 frames

df_melted = df.reset_index().melt(
    id_vars="index",
    value_vars=["Graham Scan", "Jarvis March", "Extreme Edges"],
    var_name="variable",
    value_name="value",
)

fig = px.line(
    df_melted,
    x="index",
    y="value",
    color="variable",
    labels={"index": "Frame", "value": "Cycles"},
    template="plotly_dark"
)
fig.show()
```

```python
df = pd.read_csv("convexhull_02.csv")
# df = df.iloc[:100] # Plot first 100 frames

df_melted = df.reset_index().melt(
    id_vars="index",
    value_vars=["Graham Scan", "Jarvis March", "Extreme Edges"],
    var_name="variable",
    value_name="value",
)

fig = px.line(
    df_melted,
    x="index",
    y="value",
    color="variable",
    labels={"index": "Points", "value": "Cycles"},
    log_y=True,
    template="plotly_dark"
)
fig.show()
```
